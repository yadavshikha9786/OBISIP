- Mean (₹456) is much higher than Median (₹135)
- This confirms the distribution is right-skewed
- Skewness value is positive (above 1), meaning 
  heavily skewed
- A few high-value transactions are pulling the 
  mean upward
- Median is a better measure of central tendency 
  here than mean

## Check Outlier

```python
sns.boxplot(y=df['Total Amount'])
plt.title('Outliers in Total Amount')
plt.show()
```

## Outliers Check with IQR 

```python
Q1 = df['Total Amount'].quantile(0.25)
Q3 = df['Total Amount'].quantile(0.75)
IQR = Q3 - Q1
print("Outliers:", df[(df['Total Amount'] < Q1 - 1.5*IQR) | (df['Total Amount'] > Q3 + 1.5*IQR)].shape[0])
```

### OBSERVATION:
- No statistical outliers were found in the Total Amount column using the IQR method.
- The data ranges from ₹25 to ₹2000, which falls within the IQR bounds of –1200 to 2160.
- However, notice that the mean (₹456) is much higher than the median (₹135) — this indicates the distribution is right-skewed, meaning a few high-- - - value transactions are pulling the average up.
- The data is not outlier-prone but is skewed — most customers spend low amounts, while a small number of transactions are high-value.
- This is normal behaviour in retail datasets and does not require outlier removal.


## Time Based Analysis

```python
df['Month'] = df['Date'].dt.month
df.groupby('Month')['Total Amount'].sum().plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

### OBSERVATION:
- Sales fluctuate across months with no strong seasonal pattern
- Some months show notably higher revenue — worth investigating further
- Month-wise analysis helps identify peak shopping periods
